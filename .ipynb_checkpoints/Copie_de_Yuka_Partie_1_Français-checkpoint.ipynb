# Mission 3

Crée une fonction qui prend en paramètre un dataframe, et affiche le taux de remplissage de chacune de ces colonnes sous forme de pourcentage. C'est à dire :
- colonne 1 : 73%
- colonne 2 : 99%

```python
#test
df = 100 * (yuka.isnull().sum()/len(yuka.index)).sort_values(ascending = False)
df.head(10)
```

```python
def nan(df):
  compte=1
  for col in df:
    tauxNan = round((df[col].isnull().sum()/yuka.shape[0])*100, 2)
    print("col {} {} : {} %".format(compte,col,tauxNan))      
    compte+=1                                                  
nan(yuka)
```

# Mission 4

Quel est le nombre de colonne appartenant à chaque quartile des taux de remplissage ? Autrement dit, combien de colonne sont remplies à 0%, entre 0 et 25%, entre 25 et 50%...

```python
def qt(df):
  qt0, qt1, qt2, qt3, qt4 = [],[],[],[],[]
  for col in df:
    tauxNan = round((df[col].isnull().sum()/yuka.shape[0])*100, 2)
    if tauxNan == 0:
      qt0.append(tauxNan)
    elif tauxNan < 25:
      qt1.append(tauxNan)
    elif 25 < tauxNan < 50:
      qt2.append(tauxNan)
    elif 50 < tauxNan < 75:
      qt3.append(tauxNan)
    else:
      qt4.append(tauxNan)
  print("Nombre de colonnes par quartiles:\n100% données non nulles : {}\n75-100% données non nulles : {}\n50-75% données non nulles : {}\n25-50% données non nulles : {}\n0-25% données non nulles : {}".format(len(qt0),len(qt1),len(qt2),len(qt3),len(qt4)))
                            
qt(yuka)
```
